```python
image_dir = os.path.join("..", "..", "..", "media", "FS2", "data", "pretraining_data", "png")
images = []
corrupt_files = []
for subdir, dirs, files in os.walk(image_dir):
    for file in tqdm(files):
        image = Image.open(os.path.join(image_dir, file)).convert('RGB')
        np_original = np.asarray(image)
        condition = np.fabs(np_original - 0) < 0.001
        if (np.count_nonzero(condition)/(np.count_nonzero(np_original)+1)) > 0.1: # if more than 10% of image is black
            corrupt_files.append(file)
```

```python
len(corrupt_files)
```

```python
for file in tqdm(corrupt_files):
    os.remove(os.path.join(image_dir, file))
```
